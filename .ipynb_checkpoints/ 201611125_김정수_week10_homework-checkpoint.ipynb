```python
import pyspark
import os
from pyspark.sql.types import *
from pyspark.sql.functions import udf
import numpy as np
from pyspark.sql import functions as F
from scipy.stats import norm
```

```python
myConf = pyspark.SparkConf()
spark = pyspark.sql.SparkSession.builder\
        .master("local")\
        .appName("myApp")\
        .config(conf = myConf)\
        .getOrCreate()
```

## 데이터 불러오기 & 전처리

```python
# 파일 읽어서 RDD 생성
rdd = spark.sparkContext.textFile(os.path.join("data", "제74주년 경찰의 날 기념식 축사.txt"))
```

```python
# 토크나이징
remove_null = rdd.flatMap(lambda x: x.split())\
                .map(lambda x: x.replace(",", ""))\
                .map(lambda x: x.replace(".", ""))
```

```python
_test = remove_null.collect()
```

```python
# 형태소 분석기 khaiii 사용 
from khaiii import KhaiiiApi
k_api = KhaiiiApi()

def take_khaiii(x):
    value = k_api.analyze(x)
    for word in value:
        morphs = [(mo.lex, mo.tag) for mo in word.morphs]
    return morphs
```

```python
text_analyzed = []
for _t in _test: text_analyzed += take_khaiii(_t)
```

```python
# 형태소 분석을 마친 후 다시 RDD로 변환
rdd_analyzed = spark.sparkContext.parallelize(text_analyzed)
```

```python
# 품사를 기반으로 불용어 제거
# S를 포함하는 품사 -> 특수 문자, 숫자, 한자 등등 불필요한 문자 제거
# E를 포함하는 품사 -> 선어말, 종결, 연결.. 어미 불필요한 문자 제거
# J를 포함하는 품사 -> 조사 불필요한 문자 제거
# V를 포함하는 품사 -> 지정사, 용언 불필요한 문자 제거
# 글자 수가 1개인 단어 제거 -> 의미 파악 불가한 불필요한 문자 제거
except_stopwords = rdd_analyzed.filter(lambda x: "S" not in x[1])\
                                .filter(lambda x: "E" not in x[1])\
                                .filter(lambda x: "J" not in x[1])\
                                .filter(lambda x: "V" not in x[1])\
                                .filter(lambda x: len(x[0]) != 1)
```

```python
# dataframe을 생성하기 위한 형태 변환
# 단어 / 빈도 칼럼을 가진 데이터 프레임 생성 예정
change_form = except_stopwords.map(lambda x: (x[0], 1))\
                                .reduceByKey(lambda x,y: x + y)\
                                .sortBy(lambda x: x[1])
```

```python
# 데이터프레임으로 변경 후 칼럼명 변경
df = change_form.toDF()
df = df.withColumnRenamed("_1", "words").withColumnRenamed("_2", "count")
```

```python
# tf-idf 계산을 하기 위해서는 1.단어의 빈도와 2.문서 군 내 등장 빈도가 필요
# 1은 count 칼럼에 이미 존재 
# 2를 계산하기 위한 작업 필요
```

```python
to_idf = rdd.filter(lambda x: x != "")
to_idf = to_idf.collect()
```

```python
# document frequency 계산을 위한 udf 생성
def calculate_df(word):
    cnt = 0
    for sen in to_idf:
        if word in sen: cnt += 1
    return cnt

udf_cal_df = udf(calculate_df, IntegerType())
```

```python
df = df.withColumn("df_count", udf_cal_df(df['words']))
```

```python
역문서 빈도는 한 단어가 문서 집합 전체에서 얼마나 공통적으로 나타나는지를 나타내는 값이다. 
전체 문서의 수를 해당 단어를 포함한 문서의 수로 나눈 뒤 로그를 취하여 얻을 수 있다.
```

```python
import math
```

```python
# inverse document frequency 계산을 위한 udf 생성
def calculate_idf(cnt):
    idf = math.log(cnt/len(to_idf))
    return idf

udf_cal_idf = udf(calculate_idf, DoubleType())
```

```python
df = df.withColumn("IDF", udf_cal_idf(df['df_count']))
```

```python
def tf_idf(tf, idf):
    return tf * idf

udf_cal_tfidf = udf(tf_idf, DoubleType())
```

```python
df = df.withColumn("TF-IDF", udf_cal_tfidf(*df['count', 'IDF']))
```
